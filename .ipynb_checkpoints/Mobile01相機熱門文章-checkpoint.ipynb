```python
import requests
from bs4 import BeautifulSoup
import json
url = "http://www.mobile01.com/"
res=requests.get(url + "hottopics.php?id=5",verify=False)
soup = BeautifulSoup(res.text,'lxml')
res = soup.select('tr')

for i in res:
    article_dict = {}
    try:
        article_dict['主題'] = i.select_one('.topic_gen').text.strip() ## 主題
        article_dict['主題網址']  = url + i.select_one('.subject-text').select_one('a')['href'].strip() ##主題網址
        article_dict['回覆']  = i.select_one('.reply').text.strip() ## 回覆
        article_dict['作者發文時間']  = i.select_one('.authur').select_one('p').text.strip() ## 作者發文時間
        article_dict['作者帳號']  = i.select_one('.authur').select('p')[1].text.strip() ## 作者帳號
        article_dict['最新回應時間']  = i.select_one('.latestreply').select_one('p').text.strip() ## 最新回應時間
        article_dict['最新回應帳號']  = i.select_one('.latestreply').select('p')[1].text.strip()  ## 最新回應帳號
        article_dict['相機種類連結']  = url + i.select_one('.subject-description').select_one('a')['href'].strip() ##相機種類連結
        for j in article_dict:
            print article_dict[j]     
    except Exception as e:
        print e
    print "==========article_dict==========="


```
